```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

var = ['Gender','Stress level']
  
for i in var:
    df[i] = le.fit_transform(df[i]).astype(int)
```
